```python
###matching for Reflex and XMM-Reflex is b, XMM is a
# Start time
t3_0 = time.time()

dist_XMM = Distance(z=z_XMM,cosmology=WMAP7,unit=u.Mpc)
dist_RX = Distance(z=z_RX,cosmology=WMAP7,unit=u.Mpc)
#print(distance)
# Convert ra, dec to SkyCoord
#ca = SkyCoord(ra_a*u.deg, dec_a*u.deg)
#cb = SkyCoord(ra_b*u.deg, dec_b*u.deg)

ca = SkyCoord(ra_a*u.deg, dec_a*u.deg,distance=dist_XMM,frame="icrs")
cb = SkyCoord(ra_b*u.deg, dec_b*u.deg,distance=dist_RX,frame="icrs")

# Matching
idx, d2d, d3d = ca.match_to_catalog_sky(cb,nthneighbor=2)
matches = cb[idx]

#print(reflex_name[idx],XMM_name[idx])
dra, ddec = ca.spherical_offsets_to(matches)

# Set a matching radius
idx_new = idx[d2d < 1.0 * u.arcsec]

# End time
t3_1 = time.time()
dt3 = t3_1 - t3_0
```

```python
print('Time for matching all targets:', round(dt3, 2), 'seconds')
```

```python
#matching for Reflex and ACCEPT 2.0, Reflex is b, A2 is c
#cosmology.set_current(WMAP7)
# Start time
t4_0 = time.time()
dist_A2 = Distance(z=z_A2,cosmology=WMAP7,unit=u.Mpc)
dist_RX = Distance(z=z_RX,cosmology=WMAP7,unit=u.Mpc)
#print(distance)
# Convert ra, dec to SkyCoord
#ca = SkyCoord(ra_a*u.deg, dec_a*u.deg)
#cb = SkyCoord(ra_b*u.deg, dec_b*u.deg)

cc = SkyCoord(ra_c, dec_c,distance=dist_A2,frame="icrs",unit=(u.hourangle, u.deg))
cb = SkyCoord(ra_b*u.deg, dec_b*u.deg,distance=dist_RX,frame="icrs")

# Matching
idx, d2d, d3d = cb.match_to_catalog_sky(cc)
matches = cc[idx]

#print(reflex_name[idx],XMM_name[idx])
dra, ddec = cb.spherical_offsets_to(matches)

# Set a matching radius
idx_new = idx[d2d < 10. * u.arcsec]

# End time
t4_1 = time.time()
dt4 = t4_1 - t4_0
```

```python
print('Time for matching all targets:', round(dt4, 2), 'seconds')
```
